### NL = 39 FONDO DE EMERGENCIA

```python
# --- SIMULADOR DE IA: EVALUADOR DE ESTABILIDAD FINANCIERA ---

# Inicialización de la variable de control para el bucle interactivo [7]
continuar_programa = "si"

# Uso de bucle while para mantener el programa en ejecución [5, 8]
while continuar_programa == "si":
    print("\n--- Bienvenido a su Asesor Financiero IA ---")
    print("Este sistema experto evaluará la estabilidad de su fondo de reserva.")

    # Entrada de datos: Captura de información básica [9, 10]
    gastos_mensuales = float(input("Ingrese el total de sus gastos mensuales esenciales: "))
    ahorros_actuales = float(input("¿Cuánto dinero tiene ahorrado actualmente para emergencias?: "))

    # Motor de Inferencia: Cálculo de meses de cobertura
    if gastos_mensuales > 0:
        meses_cubiertos = ahorros_actuales / gastos_mensuales

        # Aplicación de reglas lógicas predefinidas ("Si-Entonces") [4, 6, 11]
        if meses_cubiertos < 3:
            estabilidad = "MALA"
            analisis = "Su fondo es insuficiente. Se recomienda cubrir al menos 3 meses."
        elif meses_cubiertos >= 3 and meses_cubiertos <= 6:
            estabilidad = "MEDIA"
            analisis = "Tiene una base sólida, pero podría mejorar ante riesgos altos."
        else:
            estabilidad = "ALTA"
            analisis = "Su fondo es excelente. Posee una gran protección financiera."

        # Salida de información con el resultado del análisis de IA [10, 12]
        print("\n--- RESULTADO DEL ANÁLISIS ---")
        print("Usted tiene ahorros para cubrir", meses_cubiertos, "meses.")
        print("Su nivel de estabilidad es:", estabilidad)
        print("Nota de la IA:", analisis)

    else:
        print("Error: Los gastos mensuales deben ser mayores a cero.")

    # Control del bucle para permitir al usuario salir o realizar otra consulta [13, 14]
    print("\n-------------------------------------------")
    continuar_programa = input("¿Desea realizar otro análisis? (si/no): ").lower()

print("Gracias por usar el sistema experto. ¡Hasta pronto!")
```
